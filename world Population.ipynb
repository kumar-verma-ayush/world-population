```python
india=df.loc['India',years]
india.plot(kind='line')
plt.title('world populations')
plt.xlabel('years')
plt.ylabel('number of public')
plt.show()
```

```python
india=df.loc['India',years]
india.plot(kind='bar',alpha=1)
plt.title('world populations in years')
plt.xlabel('years')
plt.ylabel("In Billion Dollar")
plt.show()
```
